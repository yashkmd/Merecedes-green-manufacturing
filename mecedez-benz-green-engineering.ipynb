```python
combine = pd.concat([train_data,test_data],axis=0)
combine.tail()
```

```python
combine_cat_cols = ['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']
```

```python
# Filtering the Categorical columns in the data set
combine_cat = combine[combine_cat_cols]
combine_cat
```

*Encoding categorical columns*

```python
#pd.get_dummies(train_cat,drop_first=True).shape
# One hot encoding will create 181 new columns which will significantly increse 
#the dimensionality of the dataset. So we will try to come up with something else
```

```python
top10_X2 = combine_cat['X2'].value_counts().sort_values(ascending=False).head(10)
top10_X2
#Let's take top 10 values from each columns and encode them and ignore the others
```

```python
def binary_convertor(data,n=10):
    for col in data.columns:
        topn_col = data[col].value_counts().sort_values(ascending=False).head(n).index
        for label in topn_col:
            data[col+'_'+label] = np.where(data[col]==label,1,0)
            
    return data
```

```python
combine_cat = binary_convertor(combine_cat)
combine_cat.drop(combine_cat_cols,axis=1,inplace=True)
combine_cat.head()
```

```python
#Let's drop categorical columns from the datafame and concat encoded columns
combine.drop(combine_cat_cols,axis=1,inplace=True)
combine.head()
```

```python
combine = pd.concat([combine,combine_cat],axis=1)
#y = X['y']
#X = X.drop(['ID','y'],axis=1)
train = combine[combine['type']=='train']
test = combine[combine['type']=='test']
```
